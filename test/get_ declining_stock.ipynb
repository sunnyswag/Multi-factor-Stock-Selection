### 获取庄家还未拉升的股票

```python
import numpy as np
import sys
sys.path.append("../utils/")
import compete_banker
import date_util
```

```python
start_date = "20191008"
end_date = "20191031"
stock_list = []

date_range = date_util.get_work_day_range(start_date, end_date)

for date in date_range:
    stock = compete_banker.get_declining_stock(date, ready=False)
    if stock != []:
        for sto in stock:
            stock_list.append(sto)
```

```python
stock_list = np.unique(np.array(stock_list))
stock_list
```
